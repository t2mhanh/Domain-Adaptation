```python
import numpy as np
import scipy.misc as misc
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
from mnist_usps import *
```

```python
mnist_x_new_train, mnist_x_new_test, mnist_y_new_train, mnist_y_new_test,usps_x_new_train, usps_x_new_test, usps_y_new_train, usps_y_new_test=dataset(normalisation=False, store=False,m=.1,n=.1)
```

```python
# Convert label data into onehot encoding format
# Using the one hot encoding function in Keras;  sklearn.preprocessing.OneHotEncoder could also be used as an alternative
nb_classes=10
from keras.utils import np_utils
mnist_train_y = np_utils.to_categorical(mnist_y_new_train, nb_classes)
usps_train_y = np_utils.to_categorical(usps_y_new_train, nb_classes)
mnist_test_y= np_utils.to_categorical(mnist_y_new_test, nb_classes)
usps_test_y = np_utils.to_categorical((usps_y_new_test), nb_classes)
```
